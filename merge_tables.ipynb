```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
import os
```

```python
# Чтобы перезаписать все csv файлы из папки в один файл

import os

directory = os.fsencode("data/weather_of_zip_codes_at_date")
with open('data/weather_data.csv', 'a') as all_data_f:
    all_data_f.write(",zip_code,date,time,temperature,humidity,wind_speed\n")
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if filename.endswith(".csv"):
            with open(f'data/weather_of_zip_codes_at_date/{filename}', 'r') as one_data_f:
                next(one_data_f)
                all_data_f.write(one_data_f.read())
            continue
        else:
            continue
```

```python
df1 = pd.read_csv("data/weather_data.csv")
df1.drop('Unnamed: 0', axis=1, inplace=True)
df1
```

```python
df2 = pd.read_csv("data/region_info.csv")
df2.drop('Unnamed: 0', axis=1, inplace=True)
df2.drop('Unnamed: 0.1', axis=1, inplace=True)
df2.drop('Unnamed: 0.2', axis=1, inplace=True)
df2.drop("№", axis=1, inplace=True)
df2
```

```python
df2['hex_id'].unique()
```

```python
df_zip_code = df2.merge(df1, how="inner", on="zip_code")
df_zip_code
# df_zip_code.dropna()
```

```python
df = df_zip_code.copy()
```

```python
df['date'] = pd.to_datetime(df['date']).dt.date
df['time'] = pd.to_datetime(df['time'], format='%H:%M').dt.time
df['datetime'] = pd.to_datetime(df['date'].astype(str) + ' ' + df['time'].astype(str))
df['datetime_floor'] = df['datetime'].dt.floor('h')
df = df.drop_duplicates(subset=['hex_id', 'datetime_floor'], keep='first')
df
```

```python
full_range = pd.date_range(start=df['datetime_floor'].min(), end=df['datetime_floor'].max(), freq='h')
full_df = pd.DataFrame(full_range, columns=['datetime_floor'])
full_df

# # Заполнение пропущенных значений данными из предыдущих строк
# merged_df.fillna(method='ffill', inplace=True)
# 
# print(merged_df)
```

```python
# Объединение полного диапазона с оригинальным DataFrame
merged_df = pd.merge(full_df, df, on='datetime_floor', how='left')
merged_df.sort_values(by=['hex_id', 'date', 'time', 'datetime_floor'], ascending=True, inplace=True)
# merged_df['datetime_floor'] = pd.to_datetime(df['date'].astype(str) + ' ' + df['time'].astype(str))
merged_df
```

```python
merged_df = merged_df.ffill()
merged_df
```

```python
df = merged_df.copy()
```

```python
df['temperature_fahrenheit'] = df['temperature']
df['temperature_celsius'] = round((df.temperature_fahrenheit - 32) / 1.8).astype(int)
df
```

```python
df['temperature'] = df['temperature_celsius']
df.drop('temperature_celsius', axis=1, inplace=True)
df.drop('temperature_fahrenheit', axis=1, inplace=True)
df
```

```python
weather_df = df.reindex(
    columns=[
        'hex_id', 'street', 'region', 'latitude', 'longitude', 'datetime', 'temperature', 'humidity', 'wind_speed'
    ]
)
weather_df
```

```python
taxi_df = pd.read_csv('data/taxi_data.csv')
taxi_df.drop('Unnamed: 0', axis=1, inplace=True)
taxi_df.drop('url', axis=1, inplace=True)
taxi_df = taxi_df[taxi_df['date'].astype(str).str.isnumeric()]
taxi_df
```

```python
month = 6
year = 2024
taxi_df['date'] =  pd.to_datetime(taxi_df['date'].apply(lambda x: f'2024-06-{x}'))
taxi_df
```

```python
taxi_df['datetime'] = pd.to_datetime(taxi_df['date'].astype(str) + ' ' + taxi_df['time'].astype(str))
taxi_df.sort_values(by=['id', 'datetime'], ascending=True, inplace=True)
taxi_df['hex_id'] = taxi_df['id']
taxi_df.drop('id', axis=1, inplace=True)
taxi_df.drop('time', axis=1, inplace=True)
taxi_df.drop('date', axis=1, inplace=True)
taxi_df
```

```python
df = pd.merge(taxi_df, weather_df, on=['hex_id', 'datetime'], how='inner')
df
```

```python
df['rides_count'] = df['value']
```

```python
df = df.reindex(
    columns=[
        'hex_id', 'street', 'region', 'latitude', 'longitude', 'datetime', 'rides_count', 'temperature', 'humidity', 'wind_speed'
    ]
)
df
```

```python
df.to_csv('data/dataset.csv', index=False)
```
